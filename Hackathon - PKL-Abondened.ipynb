```python
# Creating a list of the columns for Univariate Analysis

uni_cols=list(players.columns)
#uni_cols_to_remove=['Garage_Age','Property_Age','SalePrice','ScreenPorch','EnclosedPorch','OpenPorchSF','WoodDeckSF',
#                    'GarageArea','GrLivArea','2ndFlrSF','1stFlrSF','TotalBsmtSF','BsmtUnfSF','BsmtFinSF2','BsmtFinSF1',
#                    'YearRemodAdd','LotArea','LotFrontage']

#for each in uni_cols_to_remove:
#    uni_cols.remove(each)

# Defining the function for plotting graphs for Univariate Analysis
def univariate_dist(colname,m=12,n=8):
    plt.figure(num=None,figsize=(m,n),dpi=60,facecolor='w',edgecolor='k')
    ax=sns.countplot(colname,data=players,order = players[colname].value_counts().index)
    plt.title(f'Distribution of {colname}.')
    plt.xticks(rotation=90)
    plt.grid()
    
    for p in ax.patches:
        x=p.get_bbox().get_points()[:,0]
        y=p.get_bbox().get_points()[1,1]
        ax.annotate('{:.2f}%'.format(100.*y/len(players)), (x.mean(), y), 
                    ha='center', va='bottom',color='black',weight='bold') # set the alignment of the text
        
# Calling Univariate function for each column

for each in uni_cols:
    univariate_dist(each)
```
